```python
import matplotlib.pyplot as plt
```

```python
with open('binetp.txt', 'r') as file:
    binet_size = []
    binet_time = []
    binet_ops = []
    for line in file.readlines():
        items = line.split()
        binet_size.append(2**int(items[1]))
        binet_time.append(float(items[3]))
        binet_ops.append(int(items[4]))

with open('strassenp.txt', 'r') as file:
    strassen_size = []
    strassen_time = []
    strassen_ops = []
    for line in file.readlines():
        items = line.split()
        strassen_size.append(2**int(items[1]))
        strassen_time.append(float(items[3]))
        strassen_ops.append(int(items[4]))
    


# ax = plt.subplot()
# ax.scatter(binet_size, binet_time, marker='+', label="Punkty pomiarowe")


# ax.plot(X_predict, Y_predict, label="Krzywa 1.88e-6*x^3")
# ax.legend()
# ax.set_xlabel("Rozmiar macierzy")
# ax.set_ylabel("Czas")
# ax.set_title("Czas obliczeń dla algorytmu Bineta w zależności od rozmiaru macierzy")


fig, ax = plt.subplots(2, 2, figsize=(16, 12))

X_predict = []
Y_predict = []
for i in range(1, 512):
    X_predict.append(i)
    Y_predict.append(i**3*0.00000225)
ax[0][0].scatter(binet_size, binet_time, marker='+', label='Pomiar')
ax[0][0].plot(X_predict, Y_predict, label='x^3 * 0.00000225')
ax[0][0].set_xlabel('Rozmiar macierzy')
ax[0][0].set_ylabel('Czas [s]')
ax[0][0].set_title("Czas w zależności od rozmiaru macierzy metodą Bineta")
ax[0][0].legend()


X_predict = []
Y_predict = []
for i in range(1, 512):
    X_predict.append(i)
    Y_predict.append(i**3*2)
ax[0][1].scatter(binet_size, binet_ops, marker='+', label='Pomiar')
ax[0][1].plot(X_predict, Y_predict, label='x^3 * 2')
ax[0][1].set_xlabel('Rozmiar macierzy')
ax[0][1].set_ylabel('Liczba operacji')
ax[0][1].set_title("Liczba operacji w zależności od rozmiaru macierzy metodą Bineta")
ax[0][1].legend()


X_predict = []
Y_predict = []
for i in range(1, 512):
    X_predict.append(i)
    Y_predict.append(i**2.81 * 0.0000068)
ax[1][0].scatter(strassen_size, strassen_time, marker='+', label='Pomiar')
ax[1][0].plot(X_predict, Y_predict, label='x^2.81 * 0.0000068')
ax[1][0].set_xlabel('Rozmiar macierzy')
ax[1][0].set_ylabel('Czas [s]')
ax[1][0].set_title("Czas w zależności od rozmiaru macierzy metodą Strassena")
ax[1][0].legend()


X_predict = []
Y_predict = []
for i in range(1, 512):
    X_predict.append(i)
    Y_predict.append(i**2.81 * 2.9)
ax[1][1].scatter(strassen_size, strassen_ops, marker='+', label='Pomiar')
ax[1][1].plot(X_predict, Y_predict, label='x^2.81 * 0.0000068')
ax[1][1].set_xlabel('Rozmiar macierzy')
ax[1][1].set_ylabel('Liczba operacji')
ax[1][1].set_title("Liczba operacji w zależności od rozmiaru macierzy metodą Strassena")
ax[1][1].legend()

fig.savefig("fig.png")
```
